## Exploratory Data Analysis (EDA)

### Explore ...

1. Bagaimana tren polusi udara (PM2.5) di berbagai musim dalam setahun di Gucheng?

```python
#Membuat Musim berdasarkan Bulan
```

```python
def get_season(month):
    if month in [3, 4, 5]:
        return 'Musim semi'
    elif month in [6, 7, 8]:
        return 'Musim panas'
    elif month in [9, 10, 11]:
        return 'Musim gugur'
    else:
        return 'Musim dingin'
```

```python
#mengubungkan bulan dan musim dari kota gucheng
data_gucheng['season'] = data_gucheng['month'].apply(get_season)


seasonal_avg_pm25 = data_gucheng.groupby('season')['PM2.5'].mean().reindex(['Musim semi', 'Musim panas', 'Musim gugur', 'Musim dingin'])
```

2. Faktor Cuaca Apa yang Paling Berpengaruh terhadap PM2.5 di kota Aotizhongxin?

```python
#membuat kolom cuaca
weather_columns = ['TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
matrix = data_aoti[['PM2.5'] + weather_columns].corr()
pm25_ = matrix['PM2.5'].sort_values(ascending=False)
```

3. Perbandingan PM2.5 antar kota(stasiun)

```python
#mengabungkan data PM2.5 dari semua dataset yang ada
gabungan_pm25 = pd.DataFrame({
    'Aotizhongxin': data_aoti['PM2.5'],
    'Changping': data_changping['PM2.5'],
    'Dingling': data_dingling['PM2.5'],
    'Dongsi': data_dongsi['PM2.5'],
    'Guanyuan': data_guanyuan['PM2.5'],
    'Nongzhanguan': data_nongzhanguan['PM2.5'],
    'Gucheng': data_gucheng['PM2.5']
})
gabungan_pm25
```

4. Apakah ada pola dalam waktu jam dalam sehari yang signifikan dalam kadar PM2.5 di kota guanyuan?

```python
#membuat jam dan bulan dari data kota guan yuan
hourly_avg_pm25 = data_guanyuan.groupby('hour')['PM2.5'].mean()
monthly_avg_pm25 = data_guanyuan.groupby('month')['PM2.5'].mean()
```

## Analisis Lanjutan (Opsional)

```python
#Rule Based Grouping
# Hitung rata-rata PM2.5 untuk setiap stasiun pemantauan
station_means = data_aoti.groupby('No')['PM2.5'].mean()
station_means = pd.DataFrame(station_means)

# Tentukan fungsi untuk mengelompokkan berdasarkan rata-rata PM2.5
def group_by_pm25(pm25):
    if pm25 > 100:
        return 'Tinggi'
    elif 50 < pm25 <= 100:
        return 'Sedang'
    else:
        return 'Rendah'

# Terapkan fungsi grouping
station_means['PM2.5_Group'] = station_means['PM2.5'].apply(group_by_pm25)

# Tampilkan hasil
print(station_means)
```

Pada kode di ats dilakukan perhitungan rata-rata PM2.5 untuk setiap stasiun pemantauan. Menentukan fungsi untuk mengelompokkan berdasarkan rata-rata PM2.5, merapkan fungsi grouping, serta menampilkan hasil

```python
# Visualisasi jumlah stasiun per grup
sns.countplot(x='PM2.5_Group', data=station_means)
plt.title('Jumlah Stasiun per Kategori PM2.5')
plt.show()
```

```python
#Boxplot untuk distribusi PM2.5 per grup
sns.boxplot(x='PM2.5_Group', y='PM2.5', data=station_means)
plt.title('Sebaran PM2.5 Berdasarkan Kategori')
plt.show()
```
